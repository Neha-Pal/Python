# Create a data set and make a ***table***

```python
import pandas as pd
dataset={
    'cars':['BMW','Volvo','Thar','Marcedes'],
    'passings':[10,21,5,2]
}
table=pd.DataFrame(dataset)
print(table)
```

# Panda **Series**

```python
Weights={'day1':80,'day10':75,'day20':70,'day50':65}
table=pd.Series(Weights,index=['day1','day50'])
print(table)
```
