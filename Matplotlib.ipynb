# **Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
a_1=np.array([1,2,3,4])
a_2=np.array([100,50,25,1])
#plotting in line
plt.plot(a_1,a_2)
plt.show()
```

```python
#plotting in points
```

```python
plt.plot(a_1,a_2,'.')
plt.show()
```

```python
#using marker
plt.plot(a_1,a_2,marker='P')
```

```python
plt.plot(a_1,a_2,marker='P',mec='pink')
```

```python
plt.plot(a_1,linestyle='dashed',color='r')
```

# Multiple **lines**

```python
a_1=np.array([2,9,4,7,1])
a_2=np.array([9,7,11,8,2])

plt.plot(a_1,linestyle='dashed',color='violet')
plt.plot(a_2,color='red')
plt.show()
```

# **Label**

```python
plt.title('Practice')
plt.xlabel('height')
plt.ylabel('weight')
plt.plot(a_1,linestyle='dashed',color='violet')
plt.plot(a_2,color='red')
plt.show()
```

```python
plt.grid(linestyle='--',color='grey')
plt.plot(a_1)
plt.show()
```

# **Subplot**

```python
ar=np.array([1,2,3,4])
arr=np.array([9,3,0,4])
plt.title('Sales')
plt.subplot(2,3,1)
plt.plot(ar,arr)

ar=np.array([1,2,31,4])
arr=np.array([19,3,0,4])
plt.title('Company')
plt.subplot(2,3,2)
plt.plot(ar,arr)

ar=np.array([1,21,13,4])
arr=np.array([10,3,0,14])
plt.title('Market')
plt.subplot(2,3,3)
plt.plot(ar,arr)

ar=np.array([1,20,31,4])
arr=np.array([19,13,20,4])
plt.title('Today')
plt.subplot(2,3,4)
plt.plot(ar,arr)

ar=np.array([1,2,3,14])
arr=np.array([9,1,2,4])
plt.title('Profit')
plt.subplot(2,3,5)
plt.plot(ar,arr)

ar=np.array([1,20,31,14])
arr=np.array([19,30,20,14])
plt.title('basics')
plt.subplot(2,3,6)
plt.plot(ar,arr)
```

# **ScatterPlot**

```python
plt.scatter(arr,ar,color='red')
plt.show()
```

# Bar **Graph**

```python
b1=np.array([2,6,3,8,9])
b2=np.array(['A','B','C','D','E'])
plt.bar(b1,b2,color='hotpink')
plt.show()
```

Horizontal bar

```python
plt.barh(b2,b1,color='purple')
plt.show()
```

# **Histogram**

```python
h1=np.random.normal(10,20,100)
plt.hist(h1,color='pink')
plt.show()
```

# **pie chart**

```python
p1=np.array([30,20,40,55])
fruit=['Apple','Guava','Pineapple','Mango']
plt.pie(p1, labels = fruit)
plt.legend()
plt.show()
```

Add specific colours

```python
mycolor=['hotpink','purple','black','y']
plt.pie(p1, labels = fruit, colors=mycolor)
plt.show()
```
